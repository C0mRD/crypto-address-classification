```python
# import necessary libraries
import pandas as pd
```

```python
# Load All 3 dataset - bitcoin, ethereum, dash
btc_df = pd.read_csv("../dataset/raw_data/exp1_bitcoin_addr.csv")
eth_df = pd.read_csv("../dataset/raw_data/exp2_ethereum_addr.csv")
dash_df = pd.read_csv("../dataset/raw_data/blockchair_dash_addresses_latest.tsv", delimiter='\t')
```

```python
# Drop any duplicate values
btc_df.drop_duplicates(inplace=True)
eth_df.drop_duplicates(inplace=True)
dash_df.drop_duplicates(inplace=True)
```

```python
# Take only necessary columns
btc_df = btc_df[['address']].copy()
eth_df = eth_df[['address']].copy()
dash_df = dash_df[['address']].copy()
```

```python
# Add a new column type to act as label
btc_df["type"] = "bitcoin"
```

```python
btc_df.head()
```

```python
eth_df['type'] = 'ethereum'
dash_df['type'] = 'dash'
```

```python
# I will take less number of samples for my laptop resorce limitation
samples = 50000
```

```python
# Concatenate all dataset to one
crypto_df = pd.concat([btc_df[:samples], eth_df[:samples], dash_df[:samples]])
```

```python
crypto_df
```

```python
# Shuffle the dataframe in random order
new_crypto_df = crypto_df.sample(frac=1).reset_index(drop=True).copy()
```

```python
new_crypto_df.to_csv("../dataset/processed_data/preprocessed_dataset.csv", index=False)
```
